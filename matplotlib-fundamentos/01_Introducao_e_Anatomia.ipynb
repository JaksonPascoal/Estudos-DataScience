```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Y = np.cos(X)
```

Criando a Figura e os Eixos

```python
fig, ax = plt.subplots()
```

Plotando os Dados

```python
# Cria a figura e os eixos
fig, ax = plt.subplots()

# Plota os dados
ax.plot(X, Y)

# Adiciona os rótulos e o título
ax.set_title("Onda de Cosseno")
ax.set_xlabel("Valores do Eixo X")
ax.set_ylabel("Valores do Eixo Y")

# Exibe o gráfico
plt.show()
```

```python
# Cria uma nova figura e eixos
fig, ax = plt.subplots()

# Plota a primeira linha (cosseno)
ax.plot(X, Y, label='Cosseno')

# Plota uma segunda linha (seno)
Y2 = np.sin(X)
ax.plot(X, Y2, label='Seno')

# Adiciona título, rótulos e legenda
ax.set_title("Onda de Cosseno e Seno")
ax.set_xlabel("Valores do Eixo X")
ax.set_ylabel("Valores do Eixo Y")
ax.legend()
ax.grid(True)

# Exibe o gráfico
plt.show()
```
